```python
import pandas as pd
# https://stackoverflow.com/questions/41525911/group-by-pandas-dataframe-and-select-latest-in-each-group/41531127#41531127
```

```python
src_df=pd.read_csv("https://raw.githubusercontent.com/adamFlyn/test_rl/main/src_df.csv")
hist_df=pd.read_csv("https://raw.githubusercontent.com/adamFlyn/test_rl/main/historical_df.csv")
```

```python
#keep only rows that are newer than in hist_df
new_data = src_df[src_df["yyyy_mm"].gt(src_df["state"].map(hist_df.groupby("state")["yyyy_mm"].max()))]
new_data
```

```python
#of the new rows, keep the latest updated values
picked_rows = new_data.loc[new_data.groupby("state")["yyyy_mm"].idxmax()]
picked_rows
```

```python
#concat to hist_df
output_df = pd.concat([hist_df, picked_rows], ignore_index=True)
output_df
```
